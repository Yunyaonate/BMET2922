## 5.2.1 Simple plot #1

## 5.2.2 Simple plot #2 

```python
import matplotlib.pyplot as plt
import numpy as np

N = 10
fmtString = "g:"

f = plt.figure() #creates a blank figure with optional size and name
y = range(N)

#linspace https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
x = np.linspace(0.0, 17.5, N, endpoint=True)
plt.plot(x,y,fmtString)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Exercise 5.2.2")
plt.show()
# f.savefig("Exercise_5.2.2.jpg",dip = 300)
```

## 5.2.5 Plot data from a file

```python
import matplotlib.pyplot as plt
import numpy as np

import csv
SAMPLE_TIME = 0.02

# open file for reading
filename = "pulse.csv"
with open(filename) as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    wave = []
    pulse = []

    # append each row to data
    for row in csvReader:
        wave.append(int(row[0]))
        pulse.append(float(row[1]))

t_p = [i*SAMPLE_TIME for i in range(len(pulse))]
t_w = [j*SAMPLE_TIME for j in range(len(wave))]

f = plt.figure()

plt.subplot(2,1,1)
plt.plot(t_w,wave,"c-",linewidth=1)
plt.ylim([1600, 2400])
plt.xlabel("time(s)")
plt.ylabel('Amplitude')
plt.title("Pulse waveform")

plt.subplot(2,1,2)
plt.plot(t_p,pulse,"r--",linewidth=1)
plt.ylim([1, 1.3])
plt.xlabel("time(s)")
plt.ylabel('Pulse Rate (Hz)')
plt.show()
# f.savefig("Exercise_5.2.5.jpg",dip = 300)
```
